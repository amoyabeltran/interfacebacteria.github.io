```python
from Bio import SeqIO
import os
import pandas as pd

directorio_archivos = 'ALL_Bacterias'

# Lista para almacenar los datos
datos = []

# Iterar sobre cada archivo en el directorio
for archivo in os.listdir(directorio_archivos):
    if archivo.endswith('.gbk'):
        ruta_archivo = os.path.join(directorio_archivos, archivo)

        try:
            # Leer el archivo GenBank
            with open(ruta_archivo, 'r') as file:
                record = SeqIO.read(file, 'genbank')

                # Extraer información relevante
                accession = record.id
                definition = record.description
                taxonomy = record.annotations.get('taxonomy', ['N/A'])
                species = taxonomy[-1] if taxonomy else 'N/A'
                strain = record.annotations.get('strain', 'N/A')
                organism = record.annotations.get('organism', 'N/A')
                modification_date = record.annotations.get('date', 'N/A')
                country = record.features[0].qualifiers.get('country', ['N/A'])[0]
                isolation_source = record.features[0].qualifiers.get('isolation_source', ['N/A'])[0]
                host = record.features[0].qualifiers.get('host', ['N/A'])[0]
                sequence = str(record.seq)

                # Agregar datos a la lista
                datos.append({'Accession': accession,
                              'Description': definition,
                              'Taxonomy': taxonomy,
                              'Specie': species,
                              'Strain': strain,
                              'Organism': organism,
                              'Modification Date': modification_date,
                              'Country': country,
                              'Isolation Source': isolation_source,
                              'Host': host,
                              'Sequence': sequence})

        except ValueError as e:
            print(f"Error en el archivo {archivo}: {e}")

# Crear un DataFrame con los datos
df = pd.DataFrame(datos)

# Imprimir el DataFrame
print(df)
```
