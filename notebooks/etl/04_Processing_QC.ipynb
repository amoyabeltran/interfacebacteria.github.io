 - Se filtran los nombres de los paises, agreagando una columna nueva con su especificacion

```python
import pandas as pd
import numpy as np
import geopandas as gpd
import plotly.graph_objects as go
from ast import literal_eval
```

```python
# Read CSV file
df = pd.read_csv('8_enero_daso/df_clases_Sin_eukariota.csv')

# Supongamos que countries es la lista original
countries = df.Country
df['Country'].fillna('No especificado', inplace=True)
# Eliminar duplicados usando un conjunto
unique_countries_set = set(countries)
```

```python
# Filtrado del nombre de los paises y sus localidades

listaPaises=[]
listaOtros=[]
listaPaisesCompleto=[]
listaOtrosCompleto=[]

for i in range(len(df)):
    if ':' in str(df.Country[i]):
        temp=str(df.Country[i]).find(':')
        listaPaises.append(df.Country[i][0:temp])
        listaOtros.append(df.Country[i][temp+1:])
        listaPaisesCompleto.append(df.Country[i][0:temp])
        listaOtrosCompleto.append(df.Country[i][temp+1:])
    elif ',' in str(df.Country[i]):
        temp=str(df.Country[i]).find(',')
        listaPaises.append(df.Country[i][0:temp])
        listaOtros.append(df.Country[i][temp+1:])
        listaPaisesCompleto.append(df.Country[i][0:temp])
        listaOtrosCompleto.append(df.Country[i][temp+1:])
    else:
        listaPaisesCompleto.append(df.Country[i])
        listaOtrosCompleto.append(df.Country[i])

listaPaises
```

```python
listaPaises=np.array(listaPaises)
listaOtros=np.array(listaOtros)
listaPaisesCompleto=np.array(listaPaisesCompleto)
listaOtrosCompleto=np.array(listaOtrosCompleto)

paisesUnicos=np.unique(listaPaises)
otrosUnicos=np.unique(listaOtros)
```

```python
paisesUnicos
```

```python
# Actualizacion de los paises y localidades

df['Country']=listaPaisesCompleto
df['Localidad']=listaOtrosCompleto
```

```python
# Definicion de columna nueva para el DF
#df['Longitud secuencia']=[i for i in range(len(df))]
df['Longitud']=[i for i in range(len(df))]
df['Latitud']=[i for i in range(len(df))]
```

```python
# Obtencion del largo de las secuencias
tempSequence=[]
for i in range(len(df)):
    tempSequence.append(len(df['Sequence'][i]))

df['Longitud secuencia']=tempSequence
```

```python
superkingdom=[]
phylum=[]
classes=[]
order=[]
family=[]
genus=[]

for i in range(len(df)):
    temp1=df['Taxonomy'][i]
    if len(temp1)==6 or len(temp1)==7:
        superkingdom.append(temp1[0])
        phylum.append(temp1[1])
        classes.append(temp1[2])
        order.append(temp1[3])
        family.append(temp1[4])
        genus.append(temp1[5])
    else:
        superkingdom.append('No especificado')
        phylum.append('No especificado')
        classes.append('No especificado')
        order.append('No especificado')
        family.append('No especificado')
        genus.append('No especificado')

len(superkingdom)
```

```python
df['Taxonomy:superkingdom']=superkingdom
df['Taxonomy:phylum']=phylum
df['Taxonomy:class']=classes
df['Taxonomy:order']=order
df['Taxonomy:family']=family
df['Taxonomy:genus']=genus
```
